```python
from sklearn.datasets import load_iris
import pandas as pd
dataset = load_iris()
```

```python
for key in dataset:
  print(key) 
```

```python
print(dataset)
```

```python
import pandas as pd
import numpy as np
data = pd.DataFrame(data= np.c_[dataset['data'], dataset['target']],
                     columns= dataset['feature_names'] + ['target'])
```

```python
data
```

```python
x=data.iloc[0:,0:4]
y=data.iloc[0:,4]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=77);
x_train.shape,x_test.shape,y_test,y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
from sklearn import tree
from matplotlib import pyplot as plt
feature_names=x.columns
fig=plt.figure(figsize=(25,20))
_=tree.plot_tree(clf,feature_names=feature_names,class_names={0:'setosa',1:'versicolor',2:'virginica'},filled=True,fontsize=12)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
x = dataset['data'];
import pandas as pd
data = pd.DataFrame(x, columns=dataset.feature_names)
data
```

```python
y= dataset['target']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
        data, y, test_size=0.30, shuffle=True,random_state=77)
x_train.shape,x_test.shape,y_test,y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn import tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=dataset.feature_names,  
                   class_names=dataset.target_names,
                   filled=True)
```
